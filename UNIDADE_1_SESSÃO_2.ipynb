```python
import sqlite3


def demonstracao_ddl():

    nome_banco = 'universidade.db'
    conn = sqlite3.connect(nome_banco)
    cursor = conn.cursor()

    print(f"--- Iniciando Gestão DDL para: {nome_banco} ---")


    cursor.execute('''
    CREATE TABLE IF NOT EXISTS Alunos (
        id_aluno INTEGER PRIMARY KEY AUTOINCREMENT,
        nome TEXT NOT NULL,
        email TEXT UNIQUE
    )
    ''')
    print("Tabela 'Alunos' criada com sucesso.")

    cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
    tabelas = cursor.fetchall()
    print(f"Tabelas identificadas na estrutura: {tabelas}")



    conn.close()

demonstracao_ddl()


script_sql = """
-- Configuração de Internacionalização (Fonte [2])
-- No MySQL:
-- ALTER DATABASE universidade CHARACTER SET utf8 COLLATE utf8_general_ci;

CREATE TABLE IF NOT EXISTS Cursos (
    id_curso INTEGER PRIMARY KEY,
    nome_curso TEXT
);
"""
print("\nExemplo de Script SQL organizado:")
print(script_sql)
```
