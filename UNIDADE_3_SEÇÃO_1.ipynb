```python
import sqlite3
import pandas as pd


conn = sqlite3.connect(':memory:')
cursor = conn.cursor()


cursor.execute('''
CREATE TABLE localizacao (
    Id INTEGER PRIMARY KEY AUTOINCREMENT,
    Secao VARCHAR(50) NOT NULL,
    Prateleira INTEGER NOT NULL
)
''')


cursor.execute('''
CREATE TABLE jogo (
    Cod INTEGER PRIMARY KEY AUTOINCREMENT,
    Nome VARCHAR(50) NOT NULL,
    Valor DECIMAL(6, 2) NOT NULL,
    Localizacao_Id INTEGER NOT NULL,
    FOREIGN KEY (Localizacao_Id) REFERENCES localizacao(Id)
)
''')


secoes = [
    ('Guerra', 1), ('Guerra', 2),
    ('Aventura', 100), ('Aventura', 101),
    ('RPG', 150), ('RPG', 151)
]
cursor.executemany('INSERT INTO localizacao (Secao, Prateleira) VALUES (?, ?)', secoes)

jogos = [
    ('COD 3', 125.00, 1),
    ('BF 1', 150.00, 2),
    ('GOW 4', 200.00, 3),
    ('SLY', 99.00, 4),
    ('FF XV', 205.00, 5)
]
cursor.executemany('INSERT INTO jogo (Nome, Valor, Localizacao_Id) VALUES (?, ?, ?)', jogos)



print("--- FILTRO 1: Nome do Jogo e Prateleira (INNER JOIN) [6] ---")

query1 = """
SELECT jogo.Nome, localizacao.Prateleira
FROM jogo
INNER JOIN localizacao ON localizacao.Id = jogo.Localizacao_Id
"""
print(pd.read_sql_query(query1, conn), "\n")

print("--- FILTRO 2: Jogos da Seção 'Guerra' (INNER JOIN com WHERE) [7] ---")
query2 = """
SELECT jogo.Nome
FROM jogo
INNER JOIN localizacao ON localizacao.Id = jogo.Localizacao_Id
WHERE localizacao.Secao = 'Guerra'
"""
print(pd.read_sql_query(query2, conn), "\n")

print("--- FILTRO 3: Todas as Seções e Jogos (LEFT JOIN) [8] ---")


query3 = """
SELECT localizacao.Secao, jogo.Nome
FROM localizacao
LEFT JOIN jogo ON localizacao.Id = jogo.Localizacao_Id
ORDER BY jogo.Nome ASC
"""
print(pd.read_sql_query(query3, conn))

conn.close()
```
