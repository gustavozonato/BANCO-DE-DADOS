```python
import sqlite3
import pandas as pd


conexao = sqlite3.connect('meu_banco_dados.db')
cursor = conexao.cursor()

def executar_e_mostrar(query_selecao, titulo):
    """Função auxiliar para exibir o estado atual da tabela"""
    print(f"\n--- {titulo} ---")
    try:
        df = pd.read_sql_query(query_selecao, conexao)
        print(df if not df.empty else "Tabela vazia.")
    except Exception as e:
        print(f"Erro: {e} (A tabela provavelmente não existe)")


cursor.execute("DROP TABLE IF EXISTS produtos")
cursor.execute("""
CREATE TABLE produtos (
    id INTEGER PRIMARY KEY,
    nome TEXT NOT NULL,
    preco REAL
)
""")
cursor.executemany("INSERT INTO produtos (nome, preco) VALUES (?, ?)",
                   [('Notebook', 3500.0), ('Teclado', 150.0), ('Mouse', 80.0)])
conexao.commit()

executar_e_mostrar("SELECT * FROM produtos", "Tabela Criada e Populada")


cursor.execute("DELETE FROM produtos")
conexao.commit()

executar_e_mostrar("SELECT * FROM produtos", "Após Limpeza de Dados (Estrutura Mantida)")


cursor.execute("INSERT INTO produtos (nome, preco) VALUES ('Monitor', 1200.0)")
conexao.commit()


print("\nExecutando: DROP TABLE produtos...")
cursor.execute("DROP TABLE produtos")
conexao.commit()


executar_e_mostrar("SELECT * FROM produtos", "Tentativa de consulta após o DROP")


conexao.close()
print("\nProcesso finalizado.")
```
